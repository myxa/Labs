The goal is to estimate the impact of a stenosis in a left anterior descending artery on the blood flow to the myocardium.

Assumptions about the model: the artery is 4 mm wide, 15 cm long;
the flow is steady and follows Poiseuille's law;
Blood pressure: 100 mmHg arterial, 5 mmHg venous; does not change due to stenosis;
Initial flow through the artery: 200 ml/min;

How much will the flow drop if the diameter is decreased by 50%? 70%? 90%?
Using a 0D model is recommended

```python
Q =  200 * 1.67e-8# m3/sec
Pa, Pv = 100 * 133.321995, 5 * 133.321995 # Pa
viscosity = 3e-3 # Pa/s
L = 0.15 # m
r = 0.002

R1 = 8 * viscosity * L / (np.pi * r**4) # по Пуазейлю
R1
```

```python
R_total = (Pa - Pv) / Q
R_total
```

```python
R2 = R_total - R1
R2
```

```python
R50 = 8 * viscosity * L / (np.pi * (r * 0.5)**4)
q = (Pa - Pv) / (R50 + R2)
print('flow with 50% stenosis =', 200 - q*6e+7, 'ml/min')
```

```python
R70 = 8 * viscosity * L / (np.pi * (r * 0.3)**4)
q = (Pa - Pv) / (R70 + R2)
print('flow with 70% stenosis =', 200 - q*6e+7, 'ml/min')
```

```python
R90 = 8 * viscosity * L / (np.pi * (r * 0.1)**4)
q = (Pa - Pv) / (R90 + R2)
print('flow with 90% stenosis =', (Q-q)*6e+7, 'ml/min')
```
